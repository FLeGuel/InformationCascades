```python
## Information Cascades 6 agents, i rounds
## Fabrice Le Guel - University Paris-Saclay, RITM, December 2024.
import autogen
import random
import pandas as pd

temp=0.5  # Set temperature

# Initialiser un DataFrame vide avec des colonnes
df = pd.DataFrame(columns=['round', 'agent_name', 'draw_results', 'agent_prediction', 'agents_temperature'])

# The content of the urn
urn = ['blue', 'blue', 'red']
for i in range(20):
    #round=i
    # Simulation of 6 draws with replacement
    draw_results = []
    for _ in range(6):
        draw = random.choice(urn)
        draw_results.append(draw)
 
    
    config_list = autogen.config_list_from_json(
        env_or_file="OAI_CONFIG_LIST.json"
    )
    
    llm_config = {
        "config_list": config_list,
        "timeout": 120,
        "temperature": temp,
    }
    
    Agent1 = autogen.AssistantAgent(
        name="agent1",
        system_message="You are agent1, and you are participating in a game involving two urns and six agents. The rules are as follows: There are two urns, one Blue and one Red. The Blue urn contains 2 blue balls ('b') and 1 red ball ('r'), while the Red urn contains 2 red balls ('r') and 1 blue ball ('b'). At the start of the game, one urn is randomly chosen (either Blue or Red), and this information is not revealed to any of the agents. In each round, a ball is drawn from the chosen urn, and the color of this ball is shown privately to the current agent. Based on this private information, predict the color of the urn ('Blue' or 'Red'). ",
        llm_config=llm_config,
    )
    
    Agent2 = autogen.AssistantAgent(
        name="agent2",
        system_message="You are agent2, and you are participating in a game involving two urns and six agents. The rules are as follows: There are two urns, one Blue and one Red. The Blue urn contains 2 blue balls ('b') and 1 red ball ('r'), while the Red urn contains 2 red balls ('r') and 1 blue ball ('b'). At the start of the game, one urn is randomly chosen (either Blue or Red), and this information is not revealed to any of the agents. In each round, a ball is drawn from the chosen urn, and the color of this ball is shown privately to the current agent. Additionally, you know the public predictions made so far by the other agents. Based on this private information and the public predictions, predict the color of the urn ('Blue' or 'Red').  Return 'TERMINATE' when the task is done.",
        llm_config=llm_config,
        max_consecutive_auto_reply=1
    )
    
    Agent3 = autogen.AssistantAgent(
        name="agent3",
        system_message="You are agent3, and you are participating in a game involving two urns and six agents. The rules are as follows: There are two urns, one Blue and one Red. The Blue urn contains 2 blue balls ('b') and 1 red ball ('r'), while the Red urn contains 2 red balls ('r') and 1 blue ball ('b'). At the start of the game, one urn is randomly chosen (either Blue or Red), and this information is not revealed to any of the agents. In each round, a ball is drawn from the chosen urn, and the color of this ball is shown privately to the current agent. Additionally, you know the public predictions made so far by the other agents. Based on this private information and the public predictions, predict the color of the urn ('Blue' or 'Red'). Return 'TERMINATE' when the task is done.",
        llm_config=llm_config,
        max_consecutive_auto_reply=1
    )
    
    Agent4 = autogen.AssistantAgent(
        name="agent4",
        system_message="You are agent4, and you are participating in a game involving two urns and six agents. The rules are as follows: There are two urns, one Blue and one Red. The Blue urn contains 2 blue balls ('b') and 1 red ball ('r'), while the Red urn contains 2 red balls ('r') and 1 blue ball ('b'). At the start of the game, one urn is randomly chosen (either Blue or Red), and this information is not revealed to any of the agents. In each round, a ball is drawn from the chosen urn, and the color of this ball is shown privately to the current agent. Additionally, you know the public predictions made so far by the other agents. Based on this private information and the public predictions, predict the color of the urn ('Blue' or 'Red'). Return 'TERMINATE' when the task is done.",
        llm_config=llm_config,
        max_consecutive_auto_reply=1
    )
    
    Agent5 = autogen.AssistantAgent(
        name="agent5",
        system_message="You are agent5, and you are participating in a game involving two urns and six agents. The rules are as follows: There are two urns, one Blue and one Red. The Blue urn contains 2 blue balls ('b') and 1 red ball ('r'), while the Red urn contains 2 red balls ('r') and 1 blue ball ('b'). At the start of the game, one urn is randomly chosen (either Blue or Red), and this information is not revealed to any of the agents. In each round, a ball is drawn from the chosen urn, and the color of this ball is shown privately to the current agent. Additionally, you know the public predictions made so far by the other agents. Based on this private information and the public predictions, predict the color of the urn ('Blue' or 'Red').  Return 'TERMINATE' when the task is done.",
        llm_config=llm_config,
        max_consecutive_auto_reply=1
    )
    
    Agent6 = autogen.AssistantAgent(
        name="agent6",
        system_message="You are agent6, and you are participating in a game involving two urns and six agents. The rules are as follows: There are two urns, one Blue and one Red. The Blue urn contains 2 blue balls ('b') and 1 red ball ('r'), while the Red urn contains 2 red balls ('r') and 1 blue ball ('b'). At the start of the game, one urn is randomly chosen (either Blue or Red), and this information is not revealed to any of the agents. In each round, a ball is drawn from the chosen urn, and the color of this ball is shown privately to the current agent. Additionally, you know the public predictions made so far by the other agents. Based on this private information and the public predictions, predict the color of the urn ('Blue' or 'Red').  Return 'TERMINATE' when the task is done.",
        llm_config=llm_config,
        max_consecutive_auto_reply=1
    )
    
    user_proxy = autogen.UserProxyAgent(
        name="user_proxy",
        is_termination_msg=lambda msg: msg.get("content") is not None and "TERMINATE" in msg["content"],
        human_input_mode="NEVER",
        max_consecutive_auto_reply=0,
        code_execution_config=False,
        system_message="Say 'Hello'"
    )
    
    res=user_proxy.initiate_chats(
        [
            {
                "recipient": Agent1,
                "message": "You have just observed your private information: the color of the ball drawn is "+draw_results[0]+". Based on this private information, please predict the color of the urn ('Blue' or 'Red'). Please, very short answer.",
                "clear_history": True,
                "silent": False,
                #summary_method": "reflection_with_llm"
                "summary_method": "last_msg"
            },
            {
                "recipient": Agent2,
                "message": "You have just observed your private information: the color of the ball drawn is "+draw_results[1]+". Additionally, you know the public predictions made so far by the other agent. Based on this private information and the public predictions, please predict the color of the urn ('Blue' or 'Red'). Please, very short answer.",
                "clear_history": True,
                "silent": False,
                #"summary_method": "reflection_with_llm"
                "summary_method": "last_msg"
            },
            {
                "recipient": Agent3,
                "message": "You have just observed your private information: the color of the ball drawn is "+draw_results[2]+". Additionally, you know the public predictions made so far by the other agents. Based on this private information and the public predictions, please predict the color of the urn ('Blue' or 'Red'). Please, very short answer.",
                "clear_history": True,
                "silent": False,
                #"summary_method": "reflection_with_llm"
                "summary_method": "last_msg"
            },
            {
                "recipient": Agent4,
                "message": "You have just observed your private information: the color of the ball drawn is "+draw_results[3]+". Additionally, you know the public predictions made so far by the other agents. Based on this private information and the public predictions, please predict the color of the urn ('Blue' or 'Red'). Please, very short answer.",
                "clear_history": True,
                "silent": False,
                #"summary_method": "reflection_with_llm"
                "summary_method": "last_msg"
            },
            {
                "recipient": Agent5,
                "message": "You have just observed your private information: the color of the ball drawn is "+draw_results[4]+". Additionally, you know the public predictions made so far by the other agents. Based on this private information and the public predictions, please predict the color of the urn ('Blue' or 'Red'). Please, very short answer.",
                "clear_history": True,
                "silent": False,
                #"summary_method": "reflection_with_llm"
                "summary_method": "last_msg"
            },
            {
                "recipient": Agent6,
                "message": "You have just observed your private information: the color of the ball drawn is "+draw_results[5]+". Additionally, you know the public predictions made so far by the other agents. Based on this private information and the public predictions, please predict the color of the urn ('Blue' or 'Red'). Please, very short answer.",
                "clear_history": True,
                "silent": False,
                #"summary_method": "reflection_with_llm"
                "summary_method": "last_msg"
            }
        ]
    )
    
    agent_prediction = []
    agent_name=[]
    round=[]
    agents_temperature=[]
    # extract data from res
    for chat_result in res:
        for message in chat_result.chat_history:
            if message['role'] == 'user':
                agent_prediction.append(message['content'])
                agent_name.append(message['name'])
                round.append(i)
                agents_temperature.append(temp)
                break  # On ne prend que le premier message 'user' trouvé par agent

    # Boucle pour ajouter les lignes au DataFrame
    for j in range(len(agent_name)):  # Parcourir chaque agent
        new_row = {
            'round': round[j],
            'agent_name': agent_name[j],
            'draw_results': draw_results[j],
            'agent_prediction': agent_prediction[j],
            'agents_temperature': agents_temperature[j]
        }
        # Ajouter la ligne au DataFrame
        df = pd.concat([df, pd.DataFrame([new_row])], ignore_index=True)
  
  
# Nom du fichier Excel existant
file_path = "C:/Users/FLG/Documents/ChatGPTExperiments/InformationCascades_temp05_20Rounds.xlsx"

# Lire le fichier Excel existant
existing_data = pd.read_excel(file_path)
 # Combiner les données existantes avec les nouvelles
updated_data = pd.concat([existing_data, df], ignore_index=True)

# Enregistrer les données mises à jour dans le fichier Excel
updated_data.to_excel(file_path, index=False)   
#df.to_excel(r'C:\Users\FLG\Documents\ChatGPTExperiments\InformationCascades_temp0.xlsx', index=False)
```
